```python
# Load the corpus and dictionary
from gensim import corpora, models
import pyLDAvis.gensim



corpus = corpora.MmCorpus('data/footlocker_corpus.mm')
dictionary = corpora.Dictionary.load('data/footlocker_dictionary.dict')
```

```python
lda = models.LdaModel.load('data/footlocker.lda')
followers_data =  pyLDAvis.gensim.prepare(lda, corpus, dictionary)
pyLDAvis.display(followers_data)
```
